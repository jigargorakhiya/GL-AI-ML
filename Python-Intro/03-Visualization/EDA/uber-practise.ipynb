```python
import numpy as np
import pandas as pd
```

```python
uber_data = pd.read_csv("uberdrive.csv")
uber_data.head()
```

```python
uber_data.shape
```

```python
uber_data = uber_data.rename(columns={"START_DATE*":"START_DATE","END_DATE*":"END_DATE","CATEGORY*":"CATEGORY","START*":"START","STOP*":"STOP","MILES*":"MILES","PURPOSE*":"PURPOSE"})
uber_data
```

```python
uber_data.isnull().sum()
```

```python
uber_data.info()
```

```python
uber_data = uber_data[uber_data['END_DATE'].notnull()]
uber_data.shape
```

```python
uber_data['START_DATE'] = pd.to_datetime(uber_data['START_DATE'],format="%m/%d/%Y %H:%M")
uber_data['END_DATE'] = pd.to_datetime(uber_data['END_DATE'],format="%m/%d/%Y %H:%M")
```

```python
uber_data.info()
```

```python
uber_data.describe(include="all")
```

```python
#Start and Stop points are same
uber_data[ uber_data['START'] == uber_data['STOP'] ]
```

```python
#Top 10 miles from Starting Point
uber_data.groupby("START")['MILES'].sum().sort_values(ascending = False).head(10)
```
