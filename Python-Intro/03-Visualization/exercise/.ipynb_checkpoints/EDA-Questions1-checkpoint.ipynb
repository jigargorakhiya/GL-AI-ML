### 1.4 Replacing ;?' with null values in each column?


```python
cancer_df = cancer_df.replace("?",np.nan)
#cancer_df = cancer_df.convert_objects(convert_numeric = True)
#cancer_df = cancer_df.replace("?",np.NAN)
cancer_df.columns = map(lambda x: x.replace(" ","_"),cancer_df.columns)
#cancer_df = cancer_df.convert_objects(convert_numeric = True)
cancer_df.head()
```

### 1.7 Imputing the missing values from the given dataset?

```python
# we will impute the categorical variables with 0 or 1 and continuous variables with median value

#cancer_df['Number_of_sexual_partners'] = cancer_df['Number_of_sexual_partners'].fillna(cancer_df['Number_of_sexual_partners'].mean())

cancer_df['Number_of_sexual_partners'] = cancer_df['Number_of_sexual_partners'].fillna(cancer_df['Number_of_sexual_partners'].median())
```

```python
# Imputing the missing values from First sexual intercourse

cancer_df['First_sexual_intercourse'] = cancer_df['First_sexual_intercourse'].fillna(cancer_df['First_sexual_intercourse'].median())

cancer_df['First_sexual_intercourse'].isnull().any()
```

```python
# Imputing the missing values from Num of pregnancies

cancer_df['Num_of_pregnancies'] = cancer_df['Num_of_pregnancies'].fillna(cancer_df['Num_of_pregnancies'].median())
cancer_df['Num_of_pregnancies'].isnull().any()
```
